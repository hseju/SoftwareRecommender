```python
import pandas as pd
import numpy as np
import os 
```

```python
cwd = os.getcwd()

df = pd.read_excel(cwd+"/B2B Data Provider-sample1.xlsx")
```

## Preprocessing Data

```python
#printing columns
print(df.columns)
```

```python
#removing unnecessary columns
df_pro = df.drop(['Additional Pricing Info','Website','Unlimited Data Plan'], axis=1) 
df_pro.head()
```

## READ the User data

```python
df_user = pd.read_csv("user_data.csv", index_col=0)
```

```python
df_user
```

```python
#data and contract done
data = df_user['data'].item().replace("'", "").strip("[]").replace(" ","").lower().split(",")
contract = df_user['contract'].item().replace("'", "").strip("[]").replace(" ","").split(",")
data, contract
```

```python
#duration and price
duration = df_user['duration'].item().replace("'", "").strip("[]")
price = df_user['price'].item()
duration, price
```

```python

result = df_pro[df_pro['Contract'].str.contains(r'\b(?:{})\b'.format('&|'.join(contract)))].reset_index(drop=True)
result = result[result['Solutions'].str.contains(r'\b(?:{})\b'.format('&|'.join(data)))].reset_index(drop=True)
result
```

```python
if duration in df_pro.columns[1]:
    result = result[result['Starting Price - Monthly'] <= price].reset_index(drop=True)
elif duration in df_pro.columns[2]:
    result = result[result['Starting Price - Annually'] <= price].reset_index(drop=True)
else: 
    pass
```
